```python

#Exercise VII
def human_to_dog_years(human_years):
    if human_years < 0:
        return "Error: Human years cannot negative."
    elif human_years <= 2:
        return human_years * 10.5
    else:
        return 21 + (human_years - 2) * 4

def main():
    while True:
        try:
            human_years = float(input("Enter the number of human years: "))
            result = human_to_dog_years(human_years)
            if isinstance(result, str) and result.startswith("Error"):
                print(result)
            else:
                print(f"{human_years} human years is equivalent to {result} dog years.")
            cont = input("Do you want to continue? (yes/no): ")
            if cont.lower() != "yes":
                break
        except ValueError:
            print("Error: Invalid input. Please enter a number.")

if __name__ == "__main__":
    main()
```

```python
#Exercise VIII.
import math

def compute_roots(a, b, c):
    discriminant = b**2 - 4*a*c
    if discriminant < 0:
        return "The equation has no real roots."
    elif discriminant == 0:
        root = -b / (2*a)
        return f"The equation has one real root: {root}"
    else:
        root1 = (-b + math.sqrt(discriminant)) / (2*a)
        root2 = (-b - math.sqrt(discriminant)) / (2*a)
        return f"The equation has two real roots: {root1} and {root2}"

def main():
    a = float(input("Enter the value of a: "))
    while a == 0:
        print("Error: a cannot be zero.")
        a = float(input("Enter the value of a: "))
    b = float(input("Enter the value of b: "))
    c = float(input("Enter the value of c: "))
    result = compute_roots(a, b, c)
    print(result)

if __name__ == "__main__":
    main()
```

```python
#Exercise IX.

def edit_distance(s, t):
    """Recursively calculates the edit distance between two strings."""

    if len(s) == 0:
        return len(t)  # If s is empty, insert all of t
    elif len(t) == 0:
        return len(s)  # If t is empty, delete all of s
    else:
        cost = 0
        if s[-1] != t[-1]:
            cost = 1  # Cost of 1 for substitution if last chars differ

        d1 = edit_distance(s[:-1], t) + 1  # Cost of deleting last char of s
        d2 = edit_distance(s, t[:-1]) + 1  # Cost of inserting (deleting last of t)
        d3 = edit_distance(s[:-1], t[:-1]) + cost  # substituting/matching cost

        return min(d1, d2, d3)


def main():
    s = input("Enter the first string: ")
    t = input("Enter the second string: ")

    distance = edit_distance(s, t)
    print(f"The edit distance between '{s}' and '{t}' is {distance}.")


if __name__ == "__main__":
    main()
```

```python
#Exercise X.
class Triangle:
    number_of_sides = 3  #Class variable

    def __init__(self, angle1, angle2, angle3):
        self.angle1 = angle1
        self.angle2 = angle2
        self.angle3 = angle3

    def check_angles(self):
        if self.angle1 + self.angle2 + self.angle3 == 180:
            return True
        else:
            return False


my_triangle = Triangle(90, 30, 60)
print(my_triangle.number_of_sides)
print(my_triangle.check_angles())


# Example with angles that don't add up to 180
another_triangle = Triangle(45, 45, 45)  # Example of invalid angles
print(another_triangle.number_of_sides)
print(another_triangle.check_angles())
```
